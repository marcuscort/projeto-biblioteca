###A coluna "lozalizacao" diz como os itens estão organizados por tema no acervo da biblioteca. De acordo com um sistema de classicação chamado de CDU - Classificação Decimal Universal, cada código representa uma categoria:

000 a 099: Generalidades. Ciência e conhecimento.

100 a 199: Filosofia e psicologia.

200 a 299: Religião.

300 a 399: Ciências sociais.

400 a 499: Classe vaga. Provisoriamente não ocupada.

500 a 599: Matemática e ciências naturais.

600 a 699: Ciências aplicadas.

700 a 799: Belas artes.

800 a 899: Linguagem. Língua. Linguística.

900 a 999: Geografia. Biografia. História.

##Portanto, irei criar uma coluna chamada "CDU" para especificar qual tema é o livro do empréstimo.



```python
def classificar_cdu(codigo):
    if 0 <= codigo <= 99:
        return "Generalidades. Ciência e conhecimento."
    elif 100 <= codigo <= 199:
        return "Filosofia e psicologia."
    elif 200 <= codigo <= 299:
        return "Religião."
    elif 300 <= codigo <= 399:
        return "Ciências sociais."
    elif 400 <= codigo <= 499:
        return "Classe vaga. Provisoriamente não ocupada."
    elif 500 <= codigo <= 599:
        return "Matemática e ciências naturais."
    elif 600 <= codigo <= 699:
        return "Ciências aplicadas."
    elif 700 <= codigo <= 799:
        return "Belas artes."
    elif 800 <= codigo <= 899:
        return "Linguagem. Língua. Linguística."
    elif 900 <= codigo <= 999:
        return "Geografia. Biografia. História."
    else:
        return "Código fora do intervalo."

df_completo["CDU"] = df_completo["localizacao"].apply(classificar_cdu)
```

Além disso, a redução em novembro e dezembro indica um encerramento progressivo das atividades ao fim do ano. Essa análise reforça a hipótese de que os empréstimos estão diretamente associados ao ciclo acadêmico, sendo mais intensos no início dos semestres e significativamente menores durante os períodos de férias.

###A análise dos horários de empréstimo revela um padrão bem definido ao longo do dia. O movimento começa de forma tímida nas primeiras horas — com registros residuais entre 0h e 6h, sugerindo um funcionamento limitado ou quase inexistente nesse período. A partir das 7h, os números sobem rapidamente, indicando o início do horário de atendimento. O fluxo atinge seu auge entre 10h e 17h, com destaque para o horário das 16h, que registrou o maior volume de empréstimos (92.822), seguido de perto por 17h (82.122) e 18h (81.939). Esse padrão sugere que há uma forte concentração de usuários aproveitando o período da tarde, possivelmente ao final de aulas ou turnos de trabalho. Durante a noite, entre 19h e 21h, ainda há um volume expressivo de empréstimos, o que evidencia um bom aproveitamento do turno noturno. Após as 22h, o movimento cai drasticamente, com pouquíssimos registros às 23h e 0h, refletindo o encerramento do expediente ou a ausência de atendimento nesse intervalo. Esses dados indicam que o pico de atividade acontece durante a tarde, com uma movimentação constante também no período da manhã e início da noite. Essa distribuição pode ser útil para o planejamento de escalas de atendimento, otimização de recursos humanos e ajustes nos horários de funcionamento das bibliotecas, caso necessário.

```python
def gera_box_plot(dataset,x,y,titulo,subtitulo):



  sns.set_theme(style="darkgrid", palette='Blues',font_scale=1.3)
  plt.figure(figsize=(16,10))

  ax = sns.boxplot(y= y, x= x, data= dataset,color='#4171EF')


  plt.ylim(0,max(dataset[y])*1.1)
  plt.xlabel(None)
  plt.ylabel(None)

  ax.set_title(titulo+"\n",size=20,loc='left',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset[y])*1.11,fontsize=18, ha='left',color='gray')
gera_box_plot(alunos_graduacao_acervo_circulante,'ano','quantidade','Distribuição dos empréstimos mensais', 'Realizados pelos alunos de graduação no acervo circulante')
```

Vale destacar que esse recorte representa bem a estrutura geral do conjunto de dados, já que contempla os perfis de usuários mais frequentes (alunos de graduação) e o tipo de acervo mais utilizado por eles (acervo circulante).

O boxplot permite visualizar a variação na quantidade de empréstimos realizados a cada mês, ao longo dos anos.

Além da tendência geral já identificada — com queda a partir de 2013 e uma redução mais acentuada em 2019 —, os anos de 2014 e 2017 se destacam. Se desconsiderarmos os outliers (provavelmente referentes aos meses de janeiro e dezembro), é possível notar uma estabilidade no número mensal de empréstimos nesses dois anos.

No entanto, eles se diferenciam pelo comportamento em relação aos anos anteriores: em 2014, observa-se uma queda, enquanto em 2017 houve um aumento. Seria interessante investigar com a equipe das bibliotecas o que ocorreu nesses períodos específicos, para melhor compreender o contexto e as possíveis causas dessas mudanças.
